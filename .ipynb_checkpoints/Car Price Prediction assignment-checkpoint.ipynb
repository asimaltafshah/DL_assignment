# Problem Statement::

A Chinese automobile company Geely Auto aspires to enter the US market by setting up their manufacturing unit there and producing cars locally to give competition to their US and European counterparts.

They have contracted an automobile consulting company to understand the factors on which the pricing of cars depends. Specifically, they want to understand the factors affecting the pricing of cars in the American market, since those may be very different from the Chinese market. The company wants to know:

Which variables are significant in predicting the price of a car
How well those variables describe the price of a car
Based on various market surveys, the consulting firm has gathered a large data set of different types of cars across the America market.

# task::
We are required to model the price of cars with the available independent variables. It will be used by the management to understand how exactly the prices vary with the independent variables. They can accordingly manipulate the design of the cars, the business strategy etc. to meet certain price levels. Further, the model will be a good way for management to understand the pricing dynamics of a new market.

9.Prediction

# Convert non-Numerical values ito numerical using Sklearn

```python
from sklearn.preprocessing import LabelEncoder
labelencode = LabelEncoder()
data['fueltype']=labelencode.fit_transform(data['fueltype'])
data['aspiration']=labelencode.fit_transform(data['aspiration'])
data['doornumber']=labelencode.fit_transform(data['doornumber'])
data['carbody']=labelencode.fit_transform(data['carbody'])
data['drivewheel']=labelencode.fit_transform(data['drivewheel'])
data['enginelocation']=labelencode.fit_transform(data['enginelocation'])
data['fuelsystem']=labelencode.fit_transform(data['fuelsystem'])
```

```python
data.head()
```

```python
data.info()
```

```python
corrmatrix = data.corr()
top_corr_features = corrmatrix.index
plt.figure(figsize=(20,20))
g=sns.heatmap(data[top_corr_features].corr(),annot=True,cmap='RdYlGn')
```

```python
data1=data[['horsepower','boreratio','enginesize','curbweight','carwidth','carlength','wheelbase','drivewheel','price']]
```

```python
data1
```

```python
x_value=data1.drop('price',axis=1)
```

```python
x_value
```

```python
y_value=data1['price']
```

```python
y_value.values
```

## 50 percent taining data 30% test and 20% validation

```python
from sklearn.model_selection import train_test_split
x_train_val, x_test, y_train_val, y_test = train_test_split(x_value, y_value, test_size=0.30, random_state=42) 
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train_val, y_train_val, test_size=0.285, random_state=42) #0.285*0.7=0.2
```

```python
x_train.shape
```

```python
print(x_train.shape)
print(x_test.shape)
print(x_val.shape)
print(y_train.shape)
print(y_test.shape)
print(y_val.shape)
```

```python
print(x_train)
```

# Model

```python

model= Sequential()
model.add(Dense(10,activation='relu',input_shape=(x_train.shape[1],)))
model.add(Dense(8,activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(1,))
```

```python
model.compile(optimizer='rmsprop', loss='mse',metrics=['mae'])
```

```python
history=model.fit(x_train,y_train,epochs=100,validation_data=(x_val,y_val))
```

```python
history_dict=history.history
history_dict.keys()
```

```python
import matplotlib.pyplot as plt
loss=history_dict['loss']
val_loss=history_dict['val_loss']

epochs=range(1,len(loss)+1)
plt.plot(epochs,loss,'bo',label='Training loss')
plt.plot(epochs,val_loss,'b',label='validation loss')
```

Changing model

```python
result=model.evaluate(x_test,y_test)
```

```python
pred=model.predict(x_test)
pred
```
